```python
df2=df[["normalized-losses","bore","compression-ratio","price"]]
df2.head()
```

```python
import numpy as np
msk=np.random.rand(len(df2))<0.7
train_df=df2[msk]
test_df=df2[~msk]
X_train_df= train_df[["normalized-losses","bore","compression-ratio"]]
Y_train_df=train_df[["price"]]
X_test_df= test_df[["normalized-losses","bore","compression-ratio"]]
Y_test_df=test_df[["price"]]

```

```python
X_train_df=np.asarray(X_train_df)
Y_train_df=np.asarray(Y_train_df)
X_test_df=np.asarray(X_test_df)
Y_test_df=np.asarray(Y_test_df)
```

```python
from sklearn import linear_model
clf=linear_model.LinearRegression()
clf.fit(X_train_df,Y_train_df)
print(clf.coef_)
print("\n",clf.intercept_)
```

```python
Y_pred=clf.predict(X_test_df)
```
